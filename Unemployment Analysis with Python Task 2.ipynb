# Estimated Employed:

```python
# Assuming you have a DataFrame named 'df' with 'States' and 'Estimated Employed' columns
sum_employees_by_region = df.groupby('Region')['Estimated Employed'].sum().reset_index()

# Now, 'sum_employees_by_state' contains the sum of employees for each state
print(sum_employees_by_region)
```

`Interpretation:` North region has highest estimated employees during covid-19 pandamic.

```python
x = df[['Region', 'Estimated Employed']].groupby('Region').sum()
x.plot(kind='pie', autopct='%1.1f%%', subplots=True, figsize=(15, 7), title='Estimated Employed by percentage')
plt.axis('equal')  # This line ensures that the pie chart is circular.
plt.show()
```

`Interpretation:` North region has highest percentage during covid-19 pandamic.

```python
# Sort the DataFrame by the 'Estimated Employed' column in ascending order
sum_employees_by_region_sorted = sum_employees_by_region.sort_values(by='Estimated Employed')

plt.figure(figsize=(12, 6))
plt.bar(sum_employees_by_region_sorted['Region'], sum_employees_by_region_sorted['Estimated Employed'])
plt.title("Sum of Employees by region (Ascending Order)")
plt.xlabel('Region')
plt.ylabel('Sum of Employees')
plt.xticks(rotation=65)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Assuming you have a DataFrame named 'df' with 'States' and 'Estimated Employed' columns
sum_employees_by_state = df.groupby('States')['Estimated Employed'].sum().reset_index()

# Now, 'sum_employees_by_state' contains the sum of employees for each state
print(sum_employees_by_state)
```

`Interpretation:` State wise total estimated employees

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'sum_employees_by_state' with 'States' and 'Estimated Employed' columns

# Sort the DataFrame by the 'Estimated Employed' column in ascending order
sum_employees_by_state_sorted = sum_employees_by_state.sort_values(by='Estimated Employed')

plt.figure(figsize=(12, 6))
plt.bar(sum_employees_by_state_sorted['States'], sum_employees_by_state_sorted['Estimated Employed'])
plt.title("Sum of Employees by State (Ascending Order)")
plt.xlabel('State')
plt.ylabel('Sum of Employees')
plt.xticks(rotation=65)  # Rotate x-axis labels for better readability
plt.show()
```

`Interpretation:` Uttarpradesh, Maharastra, West Bengal these are top3 estimated employees by state wise visualize clearly.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'sum_employees_by_state' with 'States' and 'Estimated Employed' columns

plt.figure(figsize=(12, 6))
plt.bar(sum_employees_by_state['States'], sum_employees_by_state['Estimated Employed'].sort_values().index)
plt.title("Sum of Employees by State")
plt.xlabel('State')
plt.ylabel('Sum of Employees')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

# Labour Participation rate:

```python
x = df[['Region', 'Estimation Labour Participation Rate']].groupby('Region').sum()
x.plot(kind='pie', autopct='%1.1f%%', subplots=True, figsize=(15, 7), title='Labour Participation Rate by Region')
plt.axis('equal')  # This line ensures that the pie chart is circular.
plt.show()
```

`Interpretation:` North Region has most effected labour participation rate (%)

```python
# average Labour Participation rate by state wise:

sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))
order = df.groupby('States')['Estimation Labour Participation Rate'].mean().sort_values().index
sns.barplot(x='States', y='Estimation Labour Participation Rate', data=df, order=order)

# add axes
plt.title("Average Labour Participation Rate")
plt.xlabel('States')
plt.ylabel('Estimation Labour Participation Rate')

plt.xticks(rotation=65)  # Rotate x-axis labels for better readability
# display the plot:
plt.show()
```

`Interpretation:` Meghalaya, Tripura, Telangana are most effected labour participation rate

# Estimated Unemployment Rate:

```python
x = df[['Region', 'Estimated Unemployment Rate']].groupby('Region').sum()
x.plot(kind='pie', autopct='%1.1f%%', subplots=True, figsize=(15, 7), title='Unemployment Rate by Region')

# This line ensures that the pie chart is circular.
plt.axis('equal') 

# display the plot:
plt.show()
```

`Interpretation:` North region has highly unemployed during covid-19

```python
# Average Unemployment rate by region

sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))
order = df.groupby('Region')['Estimated Unemployment Rate'].mean().sort_values().index
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=df, order=order)

# add axes
plt.title("Average Unemployment Rate by Region")
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate')

 # Rotate x-axis labels for better readability
plt.xticks(rotation=65)  # Rotate x-axis labels for better readability

# display the plot:
plt.show()
```

`Interpretation:` North region has highly unemployed during covid-19

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='States',y='Estimated Unemployment Rate',data=df)
plt.xticks(rotation=65)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# average Unemployment rate state wise:

sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))
order = df.groupby('States')['Estimated Unemployment Rate'].mean().sort_values().index
sns.barplot(x='States', y='Estimated Unemployment Rate', data=df, order=order)

# add axes
plt.title("Average Unemployment Rate by States")
plt.xlabel('States')
plt.ylabel('Estimated Unemployment Rate')

plt.xticks(rotation=65)  # Rotate x-axis labels for better readability
# display the plot:
plt.show()
```
