# Label Encoder
In machine learning, we usually deal with datasets which contains multiple labels in one or more than one columns. These labels can be in the form of words or numbers. label encoding refers to converting the lebels into numeric form so as to convert it into the machine-readable form.

```python
# LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
df.head()
```

```python
# value counts of Species
df['Species'].value_counts()
```

# Model Training

```python
# we need to split the data for training and testing:
from sklearn.model_selection import train_test_split
# train- 70
# test - 30
X = df.drop(columns=['Species'])
Y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.30,random_state=42)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# model training:
model.fit(x_train,y_train)
```

```python
# print metric to get performance:
print('Accuracy: ', model.score(x_test, y_test)*100)
```

```python
# KNN k- nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print('Accuracy: ', model.score(x_test,y_test)*100)
```

```python
# decision tree:
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print('Accuracy: ',model.score(x_test,y_test)*100)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred = le.inverse_transform(y_pred)
y_test = le.inverse_transform(y_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.2f}")
```

```python
print("Confusion Matrix:")
print(confusion)
```
